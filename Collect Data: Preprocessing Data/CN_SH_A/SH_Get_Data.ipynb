```python
import tushare as ts
import pandas as pd
import csv
import os.path

def EditCsv(stock):
	df = ts.get_hist_data(stock, start="2013-10-01", end="2016-11-01")
	
	df.to_csv("Stock/"+stock+".csv",columns=['open','high','low','close','volume'] )
    
def get_list(stock_list_file):
	stock_list = []
	with open(stock_list_file, 'rb') as csvfile:
		spamreader = csv.reader(csvfile, delimiter=',')
		for row in spamreader:
			stock_list.append(row)
	return stock_list

if __name__ == "__main__":
	stock_list = get_list('SH_Ticker.csv')
	for stock in stock_list:
		if not os.path.isfile("Stock/"+stock[0]+".csv"):
			try:
				EditCsv(stock[0])
				print ("Finished", stock[0])
			except:
				print (stock[0]+"cannot download")
```
